```python
import pandas as pd
import numpy as np

combined_df = pd.read_csv('forecasts/combined_forecasts.csv', index_col=0)
combined_df.index = pd.to_datetime(combined_df.index)

seasons = {
    'Spring': (3, 4, 5),
    'Summer': (6, 7, 8),
    'Autumn': (9, 10, 11),
    'Winter': (12, 1, 2)
}

all_metrics = pd.DataFrame()

for season, months in seasons.items():
    seasonal_data = combined_df[combined_df.index.month.isin(months)]
    cols_to_include = [col for col in seasonal_data.columns if 'naive' in col.lower() or 'avg' in col.lower()]
    cols_to_include.append('Actual')
    seasonal_data = seasonal_data[cols_to_include]
    metrics_df = calculate_error_metrics(seasonal_data)
    metrics_df.columns = [f'{col}_{season}' for col in metrics_df.columns]
    all_metrics = pd.concat([all_metrics, metrics_df], axis=1)

final_df = all_metrics.transpose()
final_df.index.name = 'Forecast'
final_df.columns.name = 'Metric'
base_values = np.array([11.155625 , 13.703297, 17.946432, 40.063388])  
for i in range(0, len(base_values)):
    start_index = i * 4
    end_index = start_index + 4

    final_df.loc[final_df.index[start_index:end_index], 'rMAE'] = final_df['MAE'][start_index:end_index] / base_values[i]



print(final_df)
```
